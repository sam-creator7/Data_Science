```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
pd.merge(df1,df2,on='city')
```

```python
df3 = pd.DataFrame({
    "city": ["chicago","new york","orlando","Syracuse"],
    "humidity": [65,68,75,40],
})
df3
```

```python
pd.merge(df1,df3,on='city',how='outer',indicator=True)
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df5 = pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
df5
```
